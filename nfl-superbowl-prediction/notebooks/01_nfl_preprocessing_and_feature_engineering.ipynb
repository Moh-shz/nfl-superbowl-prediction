```python
# Import Libraries
import pandas as pd
import numpy as np
import json
```

```python
# Load the dataset
df = pd.read_csv("nfl_team_stats_2002-2024.csv")

# Display all the columns
pd.set_option('display.max_columns', None)
df
```

```python
# Find games where the scores are equal (draws)
draw_games = df[df['score_away'] == df['score_home']]
draw_games[['season', 'week', 'date', 'time_et', 'neutral', 'away', 'home', 'score_away', 'score_home']]
```

```python
# Remove games that ended in a draw
df = df[df['score_away'] != df['score_home']].copy()
df
```

```python
# Drop columns containing temporal or location-related information
df = df.drop(columns=['date', 'time_et', 'neutral'])
df
```

```python
# Extract Playoff Teams

# Filter the DataFrame to include only playoff games (Wildcard, Division, Conference)
playoff_df = df[df["week"].isin(["Wildcard", "Division", "Conference"])]

# Initialise empty dictionary to store results
playoff_dict = {}

# Group by season and extract unique home and away teams per season
for season, group in playoff_df.groupby("season"):
    teams = pd.concat([group["home"], group["away"]]).dropna().unique().tolist()
    playoff_dict[season] = teams

playoff_dict
```

```python
# Extract Finalists (Superbowl Teams)

# Filter the DataFrame to include only Superbowl games
superbowl_df = df[df["week"] == "Superbowl"]

# Initialise empty dictionary to store finalists per season
finalists_dict = {}

# Group by season and extract home and away teams for each Superbowl game
for season, group in superbowl_df.groupby("season"):
    teams = pd.concat([group["home"], group["away"]]).dropna().unique().tolist()
    finalists_dict[season] = teams

finalists_dict
```

```python
# Extract Champions (Superbowl Winners)

# Initialise empty dictionary to store champions
champions_dict = {}

# Loop through each Superbowl game
for _, row in superbowl_df.iterrows():
    season = row["season"]
    if row["score_home"] > row["score_away"]:
        champion = row["home"]
    else:
        champion = row["away"]
    champions_dict[season] = champion

champions_dict
```

```python
# Remove all postseason games (Wildcard, Division, Conference, Superbowl) to keep only regular season data
df = df[~df["week"].isin(["Wildcard", "Division", "Conference", "Superbowl"])].copy()

# Convert 'week' column to integer type now that only regular season weeks remain
df["week"] = df["week"].astype(int)

# Set the 'result' column to the name of the winning team based on score
df["result"] = df.apply(lambda row: row['home'] if row['score_home'] > row['score_away'] else row['away'], axis=1)

df
```

```python
# Transform a game-level dataset into a team-level format by extracting engineered features 
# For each game, generate two rows (one for each team: home and away)

def build_precise_features(df):
    rows = []

    for _, row in df.iterrows():
        for side in ["away", "home"]:
            data = {
                "team": row[side],
                "season": row["season"],
                "week": row["week"],
                
                # 1. Total Offensive Yards
                "total_yards": row[f"yards_{side}"],
                # 2. Defensive Yards Allowed (yards gained by opponent)
                "def_yards_allowed": row[f"yards_{'home' if side == 'away' else 'away'}"],
                # 3. Turnovers = fumbles + interceptions
                "turnovers": row[f"fumbles_{side}"] + row[f"interceptions_{side}"],
                # 4. Time of Possession
                "possession_time": row[f"possession_{side}"],
                # 5. Passing Yards
                "pass_yards": row[f"pass_yards_{side}"],
                # 6. Rushing Yards
                "rush_yards": row[f"rush_yards_{side}"],
                # 7. 3rd Down Conversion Rate
                "third_down_pct": row[f"third_down_comp_{side}"] / row[f"third_down_att_{side}"] if row[f"third_down_att_{side}"] > 0 else 0,
                # 8. Red Zone Efficiency Rate
                "redzone_pct": row[f"redzone_comp_{side}"] / row[f"redzone_att_{side}"] if row[f"redzone_att_{side}"] > 0 else 0,
                # 9. point differential = team score - opponent score
                "point_diff": row[f"score_{side}"] - row[f"score_{'home' if side == 'away' else 'away'}"],
                # 10. Yards Per Play
                "ypp": row[f"yards_{side}"] / row[f"plays_{side}"] if row[f"plays_{side}"] > 0 else 0,
                
                "result": 1 if row[side] == row["result"] else 0
            }
            rows.append(data)
            
    return pd.DataFrame(rows)

data = build_precise_features(df)
data
```

```python
# Mark Super Bowl champions with 1, all others as 0
data["is_champion"] = data.apply(
    lambda row: 1 if champions_dict.get(row["season"]) == row["team"] else 0,
    axis=1
)

data
```

```python
def build_seeding_data(df, data):
    # Define division membership for each team
    division_mapping = {
        'AFC East': ['Bills', 'Dolphins', 'Patriots', 'Jets'],
        'AFC North': ['Ravens', 'Bengals', 'Browns', 'Steelers'],
        'AFC South': ['Colts', 'Jaguars', 'Texans', 'Titans'],
        'AFC West': ['Broncos', 'Chiefs', 'Raiders', 'Chargers'],
        'NFC East': ['Cowboys', 'Eagles', 'Giants', 'Commanders'],
        'NFC North': ['Lions', 'Packers', 'Bears', 'Vikings'],
        'NFC South': ['Buccaneers', 'Saints', 'Falcons', 'Panthers'],
        'NFC West': ['49ers', 'Seahawks', 'Rams', 'Cardinals']
    }

    # Create a reverse lookup from team to division
    team_to_division = {team: div for div, teams in division_mapping.items() for team in teams}

    # Build seed data
    seed_data = data[['team', 'season', 'week', 'result']].copy()
    seed_data['division'] = seed_data['team'].map(team_to_division)

    # Count wins per team per season and division
    team_wins = seed_data.groupby(['season', 'team', 'division'])['result'].sum().reset_index()
    team_wins.columns = ['season', 'team', 'division', 'wins']

    # Identify tied teams (teams with max wins in their division for that season)
    team_wins['max_wins'] = team_wins.groupby(['season', 'division'])['wins'].transform('max')
    tied_teams_df = team_wins[team_wins['wins'] == team_wins['max_wins']]

    # Basic seed dataframe with winner info
    seed_df = df[["season", "week", "away", "home", "result"]].copy()

    return seed_df, tied_teams_df, team_to_division

seed_df, tied_teams_df, team_to_division = build_seeding_data(df, data)
```

```python
# Determine Division Winners with Tie-Breakers

def get_division_winners_with_tiebreak(seed_df, tied_teams_df, team_to_division):
    division_winners = {}
    for (season, division), group in tied_teams_df.groupby(["season", "division"]):
        tied_teams = group["team"].tolist()
        head_to_head_games = seed_df[
            (seed_df["season"] == season) &
            (seed_df["home"].isin(tied_teams)) &
            (seed_df["away"].isin(tied_teams))
        ]
        win_counts = head_to_head_games["result"].value_counts()
        top_teams = win_counts[win_counts == win_counts.max()].index.tolist() if not win_counts.empty else tied_teams

        if len(top_teams) > 1:
            division_games = seed_df[
                (seed_df["season"] == season) &
                (seed_df["home"].map(team_to_division) == division) &
                (seed_df["away"].map(team_to_division) == division) &
                (seed_df["home"].isin(top_teams) | seed_df["away"].isin(top_teams))
            ]
            division_win_counts = division_games["result"].value_counts()
            division_win_counts = division_win_counts[division_win_counts.index.isin(top_teams)]
            if not division_win_counts.empty:
                top_teams = division_win_counts[division_win_counts == division_win_counts.max()].index.tolist()

        division_winners.setdefault(int(season), []).extend(top_teams)
    return division_winners

division_winners = get_division_winners_with_tiebreak(seed_df, tied_teams_df, team_to_division)
division_winners
```
